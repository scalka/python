# Three step modeling process: fit, predict and evaluate (metrics)

fitting - we have data, we have toolbox of algorithms, we put them tohether and create model

## Clustering

### Data

```python
# loads dataset from sklearn
from sklearn.datasets import make_blobs
```

```python
# in real life the y is not available
X, y = make_blobs()
```

### Algorithms

```python
# import kmeans clustering algorithm
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 3)
```

### Fit

```python
# call my model and pass data 
kmeans.fit(X)
```

### Predict

```python
# use my X to predict how im gonna group this instances
# this is how it is grouped
kmeans.predict(X)
```

```python
# save predictions
cluster_labels = kmeans.predict(X)
```

### Evaluate

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# pass real values that we know
accuracy_score(y, cluster_labels)
```

```python
# grouped them together but did not label correctly
confusion_matrix(y, cluster_labels)
```

We need to adjust the score, because it groups them correctly, it is just the random naming of each group that is different

```python
from sklearn.metrics import adjusted_rand_score
```

```python
adjusted_rand_score(y, cluster_labels)
```

### EXERCISE: Fit, predict and evaluate the iris data set using any classifier, change the parameters of the classifier or the classifier and compare the metrics

```python
# Data
from sklearn.datasets import load_iris
from sklearn.cross_validation import train_test_split
```

```python
iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.naive_bayes import GaussianNB
```


Other classification algorithms: http://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html#example-classification-plot-classifier-comparison-py

```python
# fit the model
clf = SVC()
clf.fit(X_train, y_train)
```

```python
clf.predict(X_train)
```

```python
clf_predict = clf.predict(X_test)
```

```python
## evaluate
accuracy_score(y_test, clf_predict)
```

```python
confusion_matrix(y_test, clf_predict)
```

## Evaluating models

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4)
```


We fit with the training data and we evaluate with the test data!

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python

model = KNeighborsClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python

model.score(X_test, y_test)
```
